##### 1. Import the Flights dataset from Seaborn into a Pandas DataFrame named flights_df and display the first few rows of the DataFrame.


```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

flights_df = sns.load_dataset('flights')
print(flights_df.head())
```

##### 2. Create a dictionary to map the month names to numerical values.

##### 8. Calculate the average number of passengers for each year.


```python
avg_passengers_per_year = flights_df.groupby('year')['passengers'].mean()
print(avg_passengers_per_year)
```

##### 20. Visualize the distribution of passenger counts using a box plot for each year.


```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=flights_df, x='year', y='passengers')
plt.title('Distribution of Passenger Counts by Yr')
plt.show()
```
